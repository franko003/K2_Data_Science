```python
import requests

page = requests.get('https://coinmarketcap.com/')
page.status_code
```

```python
page.content
```

### Using BeautifulSoup to Parse

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
list(soup.children)
```

```python
[type(item) for item in list(soup.children)]
```

```python
html = list(soup.children)[10]
```

```python
list(html.children)
```

```python
tbody = soup.find('tbody')
```

```python
tr = tbody.find_all('tr')
```

```python
bitcoin = tr[0]
print(bitcoin.prettify)
```

```python
# Example of pulling data for one cryptocurrency
bitcoin_name = bitcoin.find(class_='currency-name-container').get_text()
bitcoin_price = bitcoin.find(class_='price').get_text()
bitcoin_volume = bitcoin.find(class_='volume').get_text()

print(bitcoin_name)
print(bitcoin_price)
print(bitcoin_volume)
```

```python
# Extract all name, price, volume data from the page for all cryptocurrencies
currency_name_tags = tbody.select(".currency-name-container")
currency_names = [curr_name.get_text() for curr_name in currency_name_tags]
currency_names
```

```python
currency_price_tags = tbody.select(".price")
currency_prices = [curr_price.get_text() for curr_price in currency_price_tags]
currency_prices
```

```python
currency_volume_tags = tbody.select(".volume")
currency_volumes = [curr_volume.get_text() for curr_volume in currency_volume_tags]
currency_volumes
```

## Combine data into a Pandas dataframe

```python
import pandas as pd
cryptocurrencies = pd.DataFrame({'Coin': currency_names,
                                 'Price': currency_prices,
                                 'Volume': currency_volumes})
cryptocurrencies
```
